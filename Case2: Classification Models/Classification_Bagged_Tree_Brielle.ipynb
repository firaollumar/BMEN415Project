## Bagged Tree for Classification (Brielle)

```python
import pandas as pd #Importing pandas library
import numpy as np #Importing numpy library
import matplotlib.pyplot as plt #Importing Matplotlib library's "pyplot"
import seaborn as sns  # Imorting Seaborn library
import os

data = pd.read_csv("Scaled_BreastCancer_Data.csv")
```

```python
data
```

```python
del data['Unnamed: 0']
data
```

```python
#assigning features to x and target to y
X = data.drop(["Target"], axis=1) 
y = data['Target']
```

```python
#Splitting the data for training and testing
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#implementing model
from sklearn.svm import SVC
from sklearn.ensemble import BaggingClassifier
from sklearn.datasets import make_classification

bagt = BaggingClassifier(n_estimators=30, max_samples=30, random_state=0).fit(x_train, y_train)
# when you change the n_estimators and max samples, it changes the accuracy
y_pred=bagt.predict(x_test)
```

```python
#score for the training set
bagt.score(x_train, y_train)*100
```

```python
#score for the testing set
bagt.score(x_test, y_test)*100
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

cm = confusion_matrix(y_test, y_pred)

accuracy_score(y_test, y_pred)

sns.heatmap(cm, annot=True)
```
