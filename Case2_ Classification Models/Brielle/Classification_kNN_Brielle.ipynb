## kNN for Classification (Brielle)

```python
import pandas as pd #Importing pandas library
import numpy as np #Importing numpy library
import matplotlib.pyplot as plt #Importing Matplotlib library's "pyplot"
import seaborn as sns  # Imorting Seaborn library
import os

data = pd.read_csv("Scaled_BreastCancer_Data.csv")
```

```python
data
```

```python
#delete first column 
del data['Unnamed: 0']
data
```

```python
#assigning features to x and target to y
X = data.drop(["Target"], axis=1) 
y = data['Target']
```

```python
#Splitting the data for training and testing
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
#Implementing the kNN model
from sklearn.neighbors import KNeighborsClassifier

# N_neighbors means the data point is labeled by the nearest "k" data points
```

```python
#finding the optimum n_estimators

from sklearn.model_selection import GridSearchCV

knn2 = KNeighborsClassifier()
#n_neibvors for testing, test range form 2 to 25
param_grid = {'n_neighbors': np.arange(2, 25)}
#testing values
knn_gscv = GridSearchCV(knn2, param_grid, cv=2)

knn_gscv.fit(x_train, y_train)
```

```python
#check best n_neighbors value
knn_gscv.best_params_
```

```python
#implementing model with best n_neighbors
knn = KNeighborsClassifier(n_neighbors=4)

knn.fit(x_train, y_train)
y_pred=knn.predict(x_test)
#training score
knn.score(x_train, y_train)*100
```

```python
#testing score
knn.score(x_test, y_test)*100
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

cm = confusion_matrix(y_test, y_pred)

accuracy_score(y_test, y_pred)

sns.heatmap(cm, annot=True)
```
